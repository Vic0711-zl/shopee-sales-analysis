```python
import pandas as pd
import tkinter as tk
from tkinter import filedialog

# ÂàõÂª∫Êñá‰ª∂ÈÄâÊã©ÂØπËØùÊ°Ü
root = tk.Tk()
root.withdraw()  # ÈöêËóè‰∏ªÁ™óÂè£
file_path = filedialog.askopenfilename(title="Select Your CSV File", filetypes=[("CSV files", "*.csv")])

# ËØªÂèñÊï∞ÊçÆ
if file_path:
    df = pd.read_csv(file_path)
    print("‚úÖ Êï∞ÊçÆÂä†ËΩΩÊàêÂäüÔºÅ")
    print(df.info())
    print(df.head())
else:
    print("‚ùå Êú™ÈÄâÊã©Êñá‰ª∂ÔºåËØ∑ÈáçÊñ∞ËøêË°åÁ®ãÂ∫èÔºÅ")
```

```python
# Â§ÑÁêÜ total_sold ÂàóÔºåÂ∞Ü "k" ËΩ¨Êç¢‰∏∫Êï∞ÂÄº
def convert_sold(value):
    if isinstance(value, str):
        value = value.replace("k", "000").replace(".", "")  # ‰æãÂ¶Ç "8.1k" -> "8100"
    try:
        return int(value)
    except:
        return None  # Â§ÑÁêÜÂºÇÂ∏∏ÂÄº

df["total_sold"] = df["total_sold"].apply(convert_sold)

# Â§ÑÁêÜ price_actual Áº∫Â§±ÂÄºÔºàÁî®ÂùáÂÄºÂ°´ÂÖÖÔºâ
df["price_actual"].fillna(df["price_actual"].mean(), inplace=True)

# Â§ÑÁêÜ item_ratingÔºàËΩ¨Êç¢‰∏∫Êï∞ÂÄºÔºâ
df["item_rating"] = pd.to_numeric(df["item_rating"], errors="coerce")

print(df[["price_actual", "total_sold", "item_rating"]].describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ÂÖàËΩ¨Êç¢Êï∞ÊçÆÁ±ªÂûã
df["total_sold"] = pd.to_numeric(df["total_sold"], errors="coerce")  # ËΩ¨Êç¢‰∏∫Êï∞ÂÄº
df = df.dropna(subset=["total_sold"])  # Âà†Èô§Êó†ÊïàÊï∞ÊçÆ
df["total_sold"] = df["total_sold"].astype(int)  # Á°Æ‰øùÊòØÊï¥Êï∞

# Ëé∑ÂèñÈîÄÈáèÊúÄÈ´òÁöÑÂâç 10 ‰∏™ÂïÜÂìÅ
top_selling_products = df.nlargest(10, "total_sold")

# ÂèØËßÜÂåñÁïÖÈîÄÂïÜÂìÅ
plt.figure(figsize=(12, 6))
sns.barplot(x=top_selling_products["total_sold"], y=top_selling_products["title"], palette="Blues_r")
plt.xlabel("Total Units Sold")
plt.ylabel("Product Title")
plt.title("Top 10 Best-Selling Products on Shopee")
plt.show()
```

```python
# ËÆ°ÁÆó‰ª∑Ê†º‰∏éÈîÄÈáèÁöÑÁõ∏ÂÖ≥ÊÄß
df_clean = df.dropna(subset=["price_actual", "total_sold"])
correlation = df_clean[["price_actual", "total_sold"]].corr().iloc[0, 1]

# ÂèØËßÜÂåñ
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_clean["price_actual"], y=df_clean["total_sold"], alpha=0.5)
plt.xscale("log")
plt.xlabel("Product Price (log scale)")
plt.ylabel("Total Units Sold")
plt.title(f"Price vs. Sales Relationship (Correlation: {correlation:.2f})")
plt.show()
```

```python
# ËÆ°ÁÆóËØÑÂàÜ‰∏éÈîÄÈáèÁöÑÁõ∏ÂÖ≥ÊÄß
df_clean = df.dropna(subset=["item_rating", "total_sold"])
correlation_rating_sales = df_clean[["item_rating", "total_sold"]].corr().iloc[0, 1]

# ÂèØËßÜÂåñ
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_clean["item_rating"], y=df_clean["total_sold"], alpha=0.5)
plt.xlabel("Product Rating")
plt.ylabel("Total Units Sold")
plt.title(f"Customer Rating vs. Sales Relationship (Correlation: {correlation_rating_sales:.2f})")
plt.show()
```
